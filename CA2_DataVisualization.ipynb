```
import pandas as pd

url = 'https://raw.githubusercontent.com/abhikolte1/Satellite-dataset/master/Database1.csv'
df = pd.read_csv(url)
df.head()
```

```
import altair as alt

bars = alt.Chart(df, height=400, width=1000).mark_bar().encode(
  x='Country/Organization of UN Registry',
  y='count(Launch Year):Q',
  
)

text = bars.mark_text(
    align='center',
    baseline='bottom',
    dx=3  # Nudges text to right so it doesn't appear on top of the bar
).encode(
    text='count(Launch Year):Q'
)

text + bars
```

```
import altair as alt

interval = alt.selection_interval()

base = alt.Chart(df).mark_bar().encode(
  y='Country/Organization of UN Registry',
  color=alt.condition(interval, 'Users', alt.value('lightgray'))
).properties(
  selection=interval
)

base.encode(x='Longitude') | base.encode(x='Latitude')
```

```
import altair as alt

interval = alt.selection_interval()

points = alt.Chart(df).mark_point().encode(
         x='Purpose',
         y='count(Purpose)',
         color=alt.condition(interval, 'Launch Year', alt.value('lightgray'))
      ).properties(
        selection=interval
)

bar = alt.Chart(df,height=300, width=800).mark_bar().encode(
         x='Country/Organization of UN Registry',
         y='count(Country/Organization of UN Registry)',
         color=alt.condition(interval, 'Launch Year', alt.value('blue'))
      ).transform_filter(interval)

points | bar
```

```
import altair as alt

interval = alt.selection_interval()

points = alt.Chart(df, width=1000, height=200).mark_circle().encode(
  x='Launch Vehicle',
  y='count(Launch Vehicle)',
  color=alt.condition(interval, 'Purpose', alt.value('lightgray'))
).properties(
  selection=interval
).interactive()



histogram = alt.Chart(df).mark_bar().encode(
  x='count(Purpose)',
  y='Purpose',
  color='Purpose'
).transform_filter(interval)

points & histogram
```

```
import altair as alt


brush = alt.selection(type='interval', resolve='global')

base = alt.Chart(df).mark_point().encode(
    y='Users',
    color=alt.condition(brush, 'Launch Year', alt.ColorValue('lightgray'))
).add_selection(
    brush
).properties(
    width=300,
    height=300
)

base.encode(x='Purpose') | base.encode(x='Detailed Purpose')
```

```

import altair as alt

base = alt.Chart(df, width=200, height=200).mark_point(filled=True).encode(
    x=alt.X('count(Contractor):Q', title='Contractor',scale = alt.Scale()),
    y=alt.Y('count(Operator/Owner):Q', title='Count of Operating Country'), 
    tooltip="Title:N"
)



# A slider filter
year_slider = alt.binding_range(min=1985, max=2018, step=1)
slider_selection = alt.selection_single(bind=year_slider, fields=['Launch Year'], name="Launch Year")


filter_year = base.add_selection(
    slider_selection
).transform_filter(
    slider_selection
).properties(title="Slider Filtering")

country = ['Argentina','Australia','Azerbaijan','Belarus','Brazil','Canada','China','Denmark','Egypt','ESA','EUMETSAT','EUMETSAT/ESA','France','Germany',
          'Germany/USA','Greece','India','Indonesia','Iran','Israel','Italy','Japan','Kazakhstan','Luxembourg','Malaysia','Mexico','Morocco',
          'Netherlands','Philippines','Poland','Russia','Saudi Arabia','Singapore','South Africa', 'South Korea','Spain','Sweden','Taiwan','Turkey',
          'UAE','United Kingdom','USA']

users = ['Civil','Commercial','Commercial/Gov/Mil','Commercial/Government','Government/Military','Military','Military/Civil','Military/Commercial',
        'Military/Government']

# A dropdown filter
country_dropdown = alt.binding_select(options=country)
country_select = alt.selection_single(fields=['Country of Contractor'], bind=country_dropdown, name="Country of Contractor")

filter_country = base.add_selection(
    country_select
).transform_filter(
    country_select
).properties(title="Dropdown Filtering")

#color changing marks
user_radio = alt.binding_radio(options=users)

user_select = alt.selection_single(fields=['Users'], bind=user_radio, name="Type of User")
user_color_condition = alt.condition(user_select,
                      alt.Color('Users:N', legend=None),
                      alt.value('lightgray'))

highlight_users = base.add_selection(
    user_select
).encode(
    color=user_color_condition
).properties(title="Radio Button Highlighting")


( filter_year | filter_country) &  (highlight_users )
```
